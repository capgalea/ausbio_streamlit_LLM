##### AusBioTech data processing

```python
import pandas as pd
import numpy as np
import chardet
```

##### Load datasets

```python
# Path to the CSV file
file_path = 'data/Advanced Therapies directory 100424.csv'

# Detect the encoding of the original file
with open(file_path, 'rb') as f:
    result = chardet.detect(f.read())

detected_encoding = result['encoding']
print(f"Detected encoding: {detected_encoding}")

# Read the CSV file with the detected encoding

bioTech_noAddress = pd.read_csv(file_path, encoding=detected_encoding)

# Replace problematic characters
bioTech_noAddress = bioTech_noAddress.applymap(lambda x: x.replace('Ã¯Â¿Â½', '') if isinstance(x, str) else x)

# Remove leading and trailing white spaces from all string columns
bioTech_noAddress = bioTech_noAddress.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
# Path to the CSV file
file_path = 'data/AusBiotech_companies.csv'

# Detect the encoding of the original file
with open(file_path, 'rb') as f:
    result = chardet.detect(f.read())

detected_encoding = result['encoding']
print(f"Detected encoding: {detected_encoding}")

# Read the CSV file with the detected encoding
bioTech_Address = pd.read_csv(file_path, encoding=detected_encoding)

# Replace problematic characters
bioTech_Address = bioTech_Address.applymap(lambda x: x.replace('Ã¯Â¿Â½', '') if isinstance(x, str) else x)

# Remove leading and trailing white spaces from all string columns
bioTech_Address = bioTech_Address.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

##### Combine datasets

```python
# Concatenate DataFrames along rows
bioTech_data = pd.concat([bioTech_noAddress, bioTech_Address], axis=0)
bioTech_data.head()
```

```python
# Display the original columns
print(bioTech_data.columns)

# Delete the 8th to 10th columns (indices 7 to 9) and the 14th column (index 13) if they exist
columns_to_drop = [col for col in list(range(7, 11)) + [13] if col < len(bioTech_data.columns)]
bioTech_data = bioTech_data.drop(bioTech_data.columns[columns_to_drop], axis=1)

# Display the remaining columns
print(bioTech_data.columns)
```

```python
# Sort the DataFrame by the 'Companies' column
bioTech_data_sorted = bioTech_data.sort_values(by='Companies')

# Display the first few rows of the sorted DataFrame
bioTech_data_sorted.head()
```

```python
# Handle special characters (optional)
# Replace problematic characters
bioTech_data = bioTech_data.applymap(lambda x: x.replace('ï¿½', '') if isinstance(x, str) else x)

# Step 4: Write the DataFrame to a CSV file with UTF-8 encoding
bioTech_data.to_csv('bioTech_data.csv', encoding='utf-8', index=False)
```
